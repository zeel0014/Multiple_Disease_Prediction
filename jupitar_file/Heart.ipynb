### Target feature balanced or imbalanced

```python
df['target'].value_counts()
```

```python
X = df.drop('target', axis=1)
Y = df['target']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=20)
```

```python
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression

regression = LogisticRegression()
regression.fit(X_train,Y_train)
```

```python
from sklearn.metrics import accuracy_score

Y_prad = regression.predict(X_test)
accuracy_score(Y_test,Y_prad)
```

```python
df.head(10)
```

```python
print(regression.predict([[58,0,0,100,248,0,0,122,0,1.0,1,0,2]]))
print(regression.predict([[61,1,0,148,203,0,1,161,0,0,2,1,3]]))
```

```python
from sklearn.tree import DecisionTreeClassifier

classifier = DecisionTreeClassifier(criterion='entropy',random_state=68)
classifier.fit(X_train,Y_train)
```

```python
Y_prad = classifier.predict(X_test)

print(accuracy_score(Y_test,Y_prad))
print(regression.predict([[58,0,0,100,248,0,0,122,0,1.0,1,0,2]]))
print(regression.predict([[61,1,0,148,203,0,1,161,0,0,2,1,3]]))
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators = 400, criterion = 'entropy',random_state=98)
model.fit(X_train,Y_train)
```

```python
Y_prad = model.predict(X_test)

print(accuracy_score(Y_test,Y_prad))
print(regression.predict([[58,0,0,100,248,0,0,122,0,1.0,1,0,2]]))
print(regression.predict([[61,1,0,148,203,0,1,161,0,0,2,1,3]]))
```
